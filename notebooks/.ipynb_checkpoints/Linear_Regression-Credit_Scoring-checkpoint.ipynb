```python
y = datos['default']
x = datos.drop(columns = 'default')
```

```python
SEED = 77
entrenamiento_x, prueba_x, entrenamiento_y, prueba_y = train_test_split(x, y, test_size=0.30, random_state=SEED)

base_entrenamiento = entrenamiento_x.shape[0]
base_prueba = prueba_x.shape[0]

print('La base de entrenamiento tiene %s elementos y la base de prueba tiene %s elementos.' %(base_entrenamiento, base_prueba))
```

```python
modelo = LogisticRegression(max_iter=1000)
modelo.fit(entrenamiento_x, entrenamiento_y)
print(modelo.score(entrenamiento_x, entrenamiento_y))
```

### Previsiones

```python
previsiones = modelo.predict(prueba_x)
previsiones
```

```python
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(prueba_y, previsiones)
exactitud = round(exactitud, 3)*100
exactitud
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix

matriz_confusion = plot_confusion_matrix(modelo, prueba_x, prueba_y, cmap='Blues', values_format = '.3g')
matriz_confusion   
```

```python
from sklearn.metrics import classification_report

print(classification_report(prueba_y, previsiones))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

prob_prevision = modelo.predict_proba(prueba_x)[:,1]

#tfp = taxa falsos positivos
#tvp = taxa verdadeiros positivos

tfp, tvp, limite = roc_curve(prueba_y, prob_prevision)
print('roc_auc', roc_auc_score(prueba_y, prob_prevision))

plt.subplots(1, figsize=(5,5))
plt.title('Curva ROC')
plt.plot(tfp,tvp)
plt.xlabel('Especifidad')
plt.ylabel('Sensibilidad')
plt.plot([0, 1], ls="--", c = 'red') #trazando línea guia puntillada roja
plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') #trazando línea guia puntillada verde
plt.show()

#area > 0.8, discriminación excelente!
```
